```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import matplotlib.image as mpimg
import matplotlib as mpl

# Import API key
from api_keys import g_key
```

```python
cases = 'COVID-19 Cases.csv'
cases_df =pd.read_csv(cases)
```

```python
cases_df = cases_df[["Country_Region","Combined_Key", "Province_State", 
"iso3", "Lat", "Long","Population_Count",
"Date", "Case_Type", "Cases", "Difference"
]]
cases_df.head()
```

```python
aze_df = cases_df.loc[cases_df["Country_Region"] == "Azerbaijan"]
aze_df = aze_df.loc[(aze_df["Case_Type"] == "Confirmed") & (aze_df["Date"] > "5/1/2020")]
aze_df = aze_df.sort_values(by="Cases", ascending=True)
plt.figure(figsize=(10,4))
tick_locations = [value for value in aze_df["Date"]]
plt.subplot()
plt.plot(aze_df["Date"], aze_df["Cases"], color="r",label="COVID-19 Cases")
plt.yscale('linear')
plt.ylim(1800,14500)
plt.title('COVID-19 CASES IN AZERBAIJAN', fontsize = 18)
plt.ylabel("Number of Cases", fontsize = 15)
plt.xlabel("Dates", fontsize = 15)
plt.grid(True)
plt.legend(loc="best")

plt.xticks(tick_locations, aze_df["Date"], rotation="vertical")
plt.show


us_df = cases_df.loc[cases_df["Combined_Key"] == "Harris, Texas, US"]
us_df = us_df.loc[(us_df["Case_Type"] == "Confirmed") & (us_df["Date"] > "5/1/2020")]
us_df = us_df.sort_values(by="Cases", ascending=True)
plt.figure(figsize=(10,4))
tick_locations = [value for value in us_df["Date"]]
plt.subplot()
plt.plot(us_df["Date"], us_df["Cases"], color="r",label="COVID-19 Cases")
plt.yscale('linear')
plt.ylim(1800,14500)
plt.title('COVID-19 CASES IN US, TX, HOUSTON, HARRIS COUNTY', fontsize = 18)
plt.ylabel("Number of Cases", fontsize = 15)
plt.xlabel("Dates", fontsize = 15)
plt.grid(True)
plt.xticks(tick_locations, us_df["Date"], rotation="vertical")
plt.legend(loc="best")

plt.show
```

```python
map_df = cases_df[["Combined_Key", "Lat", "Long","Case_Type", "Cases"]]
map_df = map_df.dropna()
map_df = map_df.loc[cases_df["Country_Region"] == "US"]
gmaps.configure(api_key=g_key)
```

```python
locations = map_df[["Lat", "Long"]].astype(float)
heat_data = map_df["Cases"].astype(float)

intensity = map_df["Cases"].max()
fig = gmaps.figure(zoom_level=2)
heat_layer = gmaps.heatmap_layer(locations, weights=heat_data, 
                                 dissipating=False, max_intensity=intensity, point_radius = 1)
fig.add_layer(heat_layer)
fig
```
